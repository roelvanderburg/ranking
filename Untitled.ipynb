```python
def request_facet_counts(rows= 100, query= '*:*', sort=''):
    """
    :param rows: amount of rows to request
    :param query: string to query
    :param sort: string to sort the request eg. sort='ShortTimeStamp desc'
    """
    
#     url = 'http://ec2-3-120-229-133.eu-central-1.compute.amazonaws.com:8080/solr' \
#     '/ACC_Logging_Slave/select?indent=on&q={}&rows={}&sort={}&start=0&wt=json'.format(query,rows,sort)

    url = 'http://ec2-3-120-229-133.eu-central-1.compute.amazonaws.com:8080/solr/ACC_Logging_Slave/' \
    'select?facet.field=SearchText&facet.mincount=200&facet=on&indent=on&q=EventID:164%20AND%20Year:2019&rows=2&wt=json'

    response = requests.get(url)

    # Transform the request into a json
    response = response.json()

    response = response['facet_counts']['facet_fields']
    
    return response
    
```

```python
def request_into_dataframe(rows= 100, query= '*:*', sort=''):
    """
    :param rows: amount of rows to request
    :param query: string to query
    :param sort: string to sort the request eg. sort='ShortTimeStamp desc'
    """
    
    url = 'http://ec2-3-120-229-133.eu-central-1.compute.amazonaws.com:8080/solr' \
    '/ACC_Logging_Slave/select?indent=on&q={}&rows={}&sort={}&start=0&wt=json'.format(query,rows,sort)

    response = requests.get(url)

    
    # Transform the request into a json
    response = response.json()

    response = response['response']['docs'][1:]

    df =  pd.DataFrame(response)
     
    return df
    
```

```python
df = request_facet_counts()
most_searched_queries = []
for idx, data in enumerate(df['SearchText']):
    if idx % 2 == 0:
        most_searched_queries.append(data)

frames = []
for query in most_searched_queries:
    try:
        frames.append(request_into_dataframe(rows=10000, query="EventID:164 AND Year:2019 AND SearchText:\"{}\"".format(query)))
    except:
        print("jammer")
        
        
frames
```
